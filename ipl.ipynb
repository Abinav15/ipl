```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
ipl=pd.read_csv('ipl data.csv')
```

```python
ipl.head()
```

```python
ipl['player_of_match'].value_counts()
```

```python
ipl['player_of_match'].value_counts()[0:12]
```

```python
plt.figure(figsize=(15,5))
plt.bar(list(ipl['player_of_match'].value_counts()[0:12].keys()),list(ipl['player_of_match'].value_counts()[0:12]))
plt.show()
```

```python
ipl['result'].value_counts()
```

```python
plt.figure(figsize=(5,7))
plt.bar(list(ipl['result'].value_counts()[0:3].keys()),list(ipl['result'].value_counts()[0:3]))
plt.show()
```

```python
ipl['toss_winner'].value_counts()
```

```python
plt.figure(figsize=(7,7))
plt.pie(list(ipl['toss_winner'].value_counts()),labels=list(ipl['toss_winner'].value_counts().keys()),autopct='%0.1f%%')
plt.show()
```

```python
plt.figure(figsize=(15,5))
plt.bar(list(ipl['toss_winner'].value_counts()[0:8].keys()),list(ipl['toss_winner'].value_counts()[0:8]))
plt.show()
```

```python
batting_first=ipl[ipl['win_by_runs']!=0]
batting_first.head()
```

```python
batting_first['winner'].value_counts()
```

```python
plt.figure(figsize=(7,7))
plt.pie(list(batting_first['winner'].value_counts()),labels=list(batting_first['winner'].value_counts().keys()),autopct='%0.1f%%')
plt.show()
```
